```python
#importing the required packages

import pandas as pd
import numpy as np
import codecs
import matplotlib.pyplot as plt
```

```python
#Get the dataset

x_train = np.load('x_train.npy',mmap_mode='r')
y_train = np.load('y_train.npy',mmap_mode='r')
x_test  = np.load('x_test.npy' ,mmap_mode='r')
y_test  = np.load('y_test.npy' ,mmap_mode='r')


x_train  = x_train.flatten()
x_test   = x_test.flatten()
y_train  = y_train.flatten()
y_test   = y_test.flatten()


print('x_train is  : \n' ,x_train)
print('the shape of x_train is :') 
print('\n')
print('y_train is  : \n',y_train)
print('the shape of x_train is :') 
print('x_test is  : \n' ,x_test)
print('the shape of x_test is :') 
print('\n')
print('y_test is  : \n',y_test)
print('the shape of x_test is : ') 
```

```python
X = x_train
Y = y_train

# calculate mean of x & y using an inbuilt numpy method mean()
mean_x = np.mean(X)
mean_y = np.mean(Y)

# total no.of input values
m = len(X)

# using the formula to calculate m & c
numer = 0
denom = 0
for i in range(m):
  numer += (X[i] - mean_x) * (Y[i] - mean_y)
  denom += (X[i] - mean_x) ** 2
m = numer / denom
c = mean_y - (m * mean_x)

print (f'm = {m} \nc = {c}')

# plotting values and regression line
max_x = np.max(X) + 10
min_x = np.min(X) - 0

# calculating line values x and y
x = np.linspace (min_x, max_x, 3)
y = c + m * x

ss_t = 0 #total sum of squares
ss_r = 0 #total sum of square of residuals

for i in range((100)): 
  y_pred = c + m * X[i]
  ss_t += (Y[i] - mean_y) ** 2
  ss_r += (Y[i] - y_pred) ** 2
r2 = 1 - (ss_r/ss_t)

print('r2 = ',r2)


plt.plot(x, y, color='black', label='Regression Line')
plt.scatter(X, Y, c='orange', label='data points')

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()

```

```python

#linear regrission with gradient decent on training data set 


class LinearRegression(object): 

     

    def __init__(self): 

        self.b0 = 0 

        self.b1 = 0 

        self.errors = [] 

        self.b0_errors = [] 

        self.b1_errors = [] 
        
        
        

    def train(self, x, y,initial_b0 = 0, initial_b1 = 0, num_itrs = 100000, alpha = .0001): 

        ''' 

        using train function to do the gradient descent with some logs 

        ''' 

        self.b0 = initial_b0 

        self.b1 = initial_b1 

        print("Starting gradient descent at b0 = {0}, b1 = {1}".format(self.b0, self.b1)) 

        print("Running...") 

        self.gradient_descent(x,y, alpha, num_itrs) 

        print("After {0} iterations b0 = {1}, b1 = {2}".format(num_itrs, self.b0, self.b1)) 

        print("errors moved from {0} to {1}".format(self.errors[0], self.errors[-1])) 

     

    def predict(self, x, b0=None, b1=None): 

        ''' 

        predicting the y using the current b0 and b1, with optional assignment for them 

         

        ''' 

        if b0 == None or b1 == None: 

            b0 = self.b0 

            b1 = self.b1 

        x = np.array(x) 

        prediction = b0 + b1 * x 

        return prediction 

     

    def compute_cost(self, x, y): 

        ''' 

        Comput cost for linear regression 

        ''' 

        #Number of training samples 

        m = len(y) 

  

        predictions = self.predict(x) 

  

        sq_errs = ((predictions - y) ** 2).sum() 

  

  

        return sq_errs/(2*m) 

  

  

    def gradient_descent(self, x, y, alpha, num_iters): 

        ''' 

        preform the gradient descent to adjust b0 and b1 

        ''' 

        self.b0 = 0 

        self.b1 = 0 

        m = len(y) 

        self.errors = [] 

        for i in range(num_iters): 

  

            predictions = self.predict(x, self.b0, self.b1) 

             

            b0_grad = (1/m) * (predictions - y).sum() 

            b1_grad = (1/m) * x.T.dot(predictions - y) 

             

            self.b0 -= alpha * b0_grad 

            self.b1 -= alpha * b1_grad 

  

            self.errors.append(self.compute_cost(x,y)) 

            self.b0_errors.append(abs(b0_grad)) 

            self.b1_errors.append(abs(b1_grad)) 

  
     

    def plot_errors(self): 

        plt.plot(self.errors) 

        plt.show() 

     

    def plot_b0_b1_errors(self): 

        fig, axes = plt.subplots(ncols=2) 

        axes[0].plot(self.b0_errors, label="b0 errors", ) 

        axes[1].plot(self.b1_errors, label="b1 errors") 

        plt.show() 
  

X = x_train 

Y = y_train 


# calculate mean of x & y using an inbuilt numpy method mean()
mean_x = np.mean(X)
mean_y = np.mean(Y)
# using the formula to calculate m & c
numer = 0
denom = 0
m = len(X)
for i in range(m):
  numer += (X[i] - mean_x) * (Y[i] - mean_y)
  denom += (X[i] - mean_x) ** 2
m = numer / denom
c = mean_y - (m * mean_x)

print (f'm = {m} \nc = {c}')



# calculating line values x and y
x = np.linspace (min_x, max_x, 100)
y = c + m * x

ss_t = 0 #total sum of squares
ss_r = 0 #total sum of square of residuals

for i in range((100)): # val_count represents the no.of input x values
  y_pred = c + m * X[i]
  ss_t += (Y[i] - mean_y) ** 2
  ss_r += (Y[i] - y_pred) ** 2
r2 = 1 - (ss_r/ss_t)

print('r2 = ',r2)





l_reg = LinearRegression() 

l_reg.train(X,Y,num_itrs=100000, alpha = .0001) 

 

 

ypred = l_reg.predict(X) 

print(ypred,'\n\n\n\n') 

 

 

for i in range(len(Y)): 

    print(Y[i], ypred[i]) 
    
    

print('\n\n\n\n')


 

plt.style.use('fivethirtyeight') 

plt.scatter(X,Y, color='orange') 

plt.plot(X,ypred) 

plt.show() 

 

 
print('\n\n\n\n')

 

l_reg.train(X,Y,num_itrs=1000000, alpha = .0001) 

ypred = l_reg.predict(X) 

plt.style.use('fivethirtyeight') 

plt.scatter(X,Y, color='orange') 

plt.plot(X,ypred,color='red' ) 

plt.show()  


print('\n\n\n\n')
 

 

l_reg.plot_b0_b1_errors() 

```
